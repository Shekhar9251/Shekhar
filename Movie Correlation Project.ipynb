```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

#loading the .csv file
df= pd.read_csv(r'D:\PROJECT\new\moviess.csv', index_col = 0)
df.head()
```

```python
#checking if there are any null values i.e. empty
df.isnull()
```

```python
#checking the number of null values in each column
np.sum(df.isnull())
```

```python
#filling the null values with zero
df2= df.fillna(value= 0)
df2
```

```python
#checking to see if all the null values i.e. empty places are filled
np.sum(df2.isnull())
```

```python
df2
```

```python
#checking the data types of each column
df2.dtypes
```

```python
#converting the budget and gross column from float to int data type for better visula
df2["budget"]= df2["budget"].astype("int64")
df2["gross"]= df2["gross"].astype("int64")
print(df2.dtypes)
df2.head()
```

```python
# the 'year' column had different years than the actual release year in the 'released' column. That means there was a mistake
# extracting the year from the 'released' column and adding it in a new column named 'correct_year'
import re
def extract_year(date_string):
    match = re.search(r'\b\d{4}\b', str(date_string))
    return int(match.group()) if match else None
df2['correct_year'] = df2['released'].apply(extract_year)
df2
```

```python
df2.dtypes
```

```python
# also changing the data type of the newly created 'correct_year' column from float to int.
# since few rows of the correct_year column have null value we cannot directly convert it to int.
# we need to fill the null places with 0
df2["correct_year"].fillna(0, inplace=True)
df2["correct_year"]= df2["correct_year"].astype("int64")
df2.dtypes
```

```python
# checking for any remaining null values
np.sum(np.sum(df2.isnull()))
```

```python
# sorting the the table in descending order with respect to gross
sorted_df2 = df2.sort_values(by='gross', ascending=False)
sorted_df2.head(200)
```

```python
# from the table I'm gussing that there is a high correlation between budget and gross
# creating a scatter plot to check correlation between budget and gross
plt.scatter(sorted_df2["budget"], sorted_df2["gross"])
plt.title("BUDGET vs GROSS")
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.show()
```

```python
# checking the 1st two values manually on the plot to see weather python was able to plot the graph correctly or not
# it's correctly plotted
sorted_df2.head(2)
```

```python
# creating a seaborn plot for better understanding of the coorelation
# still my guess is positive
sns.regplot(x="budget", y="gross", data=sorted_df2, line_kws={"color":"red"})
```

```python
#let's start looking at correlation between other Movie Features
numeric_columns = sorted_df2.select_dtypes(include="number") #Exclude non-numeric columns
numeric_columns.corr("pearson") #pearson(using this) #kendall #spearman 
```

```python
# Even after checking with pearson correlation we can see that there is a high correlation between Budget and Gross
#let's plot a heatmap for understanding the same visually
correlation_matrix= numeric_columns.corr("pearson")
sns.heatmap(correlation_matrix, annot= True)
plt.title("Correlation between different movie features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
```

```python
sorted_df2.head()
```

```python
# Now, we also need to check the correlation between the movie features like company, country, star, writer, etc. 
# But we can see that these columns are object datatypes and hence the correlation cannot be established since we need numerical value
# So, we will be giving random numbers to object datatype colunms for checking coorelation
df_numerised= sorted_df2
for col in df_numerised.columns :
    if(df_numerised[col].dtypes == "object") :
        df_numerised[col]= df_numerised[col].astype("category")
        df_numerised[col]= df_numerised[col].cat.codes
df_numerised.head(100)
    
```

```python
sorted_df2.head(100)
# we can see that a random numerical value has been assigned to the object data types
```

```python
# now checking correlation for all movie features
coorelation_matrix2= df_numerised.corr("pearson")
sns.heatmap(coorelation_matrix2, annot= True)
plt.title("Correlation between different movie features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# the heat map looks cluttered so we will look at the correlation in a tabbular form
df_numerised.corr("pearson")
```

```python
# the correlation is not very practical to read 
# using unstacking we will reduce the double correlation values to only single for better and quick reading
coorelation_matrix2= df_numerised.corr()
corr_pairs= coorelation_matrix2.unstack()
corr_pairs.tail(60)
```

```python
# still difficult and time taking to check the correaltion
# let's sort it out then
sorted_pairs= corr_pairs.sort_values()
sorted_pairs.tail(60)
```

```python
# now we were able to read the correlation easily
# but still we dont need those movie features which have low correlation
# so, lets sort it even more to only see correlation values which have correlation >0.5
high_correlation= sorted_pairs[sorted_pairs > 0.5]
high_correlation
```
